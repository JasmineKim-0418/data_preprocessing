# 1. 엑셀 파일 읽어오기

```python
#df = pd.read_excel('enterprise_db_list1.xlsx',engine='openpyxl') # 데이터프레임 형식으로 읽어옴.

# 엑셀 파일 읽기 (왜 안될까?)
#df = pd.read_excel('/content/drive/MyDrive/데이터스/enterprise_db_list1.xlsx')


# csv 파일 읽기
df = pd.read_csv('/content/drive/MyDrive/데이터스/enterprise_db_list1_수정.csv')
json_data = df.to_json(orient='records',force_ascii=False)

with open('output.json','w') as file :
  file.write(json_data)
```

#2. json 형태로 바꾸기

```python
file_path = '/content/drive/MyDrive/데이터스/output.json'

with open(file_path, 'r') as file:
    json_data = json.load(file)
```

```python
formatted_data = json.dumps(json_data, indent=4, ensure_ascii=False) # 아스키코드 그대로

print(formatted_data)
```

#3. json 전처리 - 파일 중 "No." 값의 소수점 없애기

```python
# 리스트 형식의 JSON 데이터를 문자열로 변환
json_str = json.dumps(json_data)


# JSON 문자열을 파싱하여 Python 객체로 변환
parsed_data = json.loads(json_str)

# "No." 값을 정수로 변환
for item in parsed_data:
  no_value = item["No."]
  if no_value is not None:
    item["No."] = int(no_value)

formatted_json = json.dumps(parsed_data, indent=4, ensure_ascii=False) # 아스키코드 그대로


print(formatted_json)

```

#4. json 전처리 - json 파일 순서 바꾸기

1. 업체명
2. 본사주소
3. 전화번호
4. 기업규모
5. 업종
6. 대표자명
7. 사업자번호

```python
# JSON 문자열을 파싱하여 Python 객체로 변환
parsed_data = json.loads(formatted_json)

# 필드를 재배열하여 새로운 JSON 데이터 생성
new_data = []
for item in parsed_data:
    new_item = {
        "No.": item["No."],
        "업체명": item["업체명"],
        "본사주소": item["위치"],
        "전화번호": item["전화번호"],
        #"기업규모": "",  # 기업규모 필드는 데이터에서 제공되지 않았으므로 빈 문자열로 설정
        "업종": item["분류"],
        "대표자명": item["대표자"],
        #"사업자번호" : "",
        "이메일": item["이메일"]

    }
    new_data.append(new_item)

# 새로운 JSON 데이터 출력
new_json = json.dumps(new_data, indent=4, ensure_ascii=False)
print(new_json)
```

#5. json 전처리 - Null 값 없애기



```python
# JSON 데이터 파싱
parsed_data = json.loads(new_json)

# null 값을 None으로 변경하는 함수
def replace_null_with_none(obj):
    for key, value in obj.items():
        if value == "null":
            obj[key] = None

# 모든 객체에서 null 값을 None으로 변경.
for obj in parsed_data:
    replace_null_with_none(obj)



print(parsed_data)
```

```python
# None 값을 가진 키를 제거하는 함수
def remove_none_values(parsed_data):
    for item in parsed_data:
        keys_to_remove = []
        for key, value in item.items():
            if value is None:
                keys_to_remove.append(key)
        for key in keys_to_remove:
            del item[key]

remove_none_values(parsed_data)



formatted_data = json.dumps(parsed_data, indent=4, ensure_ascii=False)
print(formatted_data)
```

```python
# JSON 데이터를 텍스트 파일로 저장
with open('formatted_data.json', 'w') as f:
    json.dump(formatted_data, f, ensure_ascii=False, indent=4)

print("파일이 성공적으로 저장되었습니다.")
```
